# Fill Null Value in specific column

```python
df6=df.fillna({

               'Delta 15 N (o/oo)':999,
               'Delta 13 N (o/oo)':666
})
```

```python
df6
```

#Filling Null for different columns with mean value of that column values

```python
df7= df.fillna({'Delta 15 N (o/oo)': df['Delta 15 N (o/oo)'].mean(),
                'Delta 13 N (o/oo)': df['Delta 15 N (o/oo)'].mean()
                })
```

```python
df7
```
